```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import tree
from sklearn.datasets import load_breast_cancer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
```

```python

#Constructing tab for metric performance using arrays and not dictionaries
#Legenda: 
# Models for different machine learning models
# Names for model names, 
# Metrics for evaluating performances


models = [
    KNeighborsClassifier(),
    KNeighborsClassifier(3),
    LogisticRegression(solver='liblinear'),
    DecisionTreeClassifier(max_depth = 2),
    BaggingClassifier(),
    RandomForestClassifier(),
    AdaBoostClassifier(),
    GradientBoostingClassifier(),
    XGBClassifier(use_label_encoder=False,eval_metric="logloss")
]

names = np.array([
    "Knn",
    "Knn_3",
    "Lr",
    "Dt",
    "Bc",
    "Rf",
    "AdaB",
    "Gb",
    "XGB"
])
metrics = np.array([
    "accuracy",
    "prec",
    "rec",
    "auc"
])
```

```python
#Custom Function for calculating precision and recall, and plotting PR and ROC curve


def Confusion_matrix_precision_and_recall(probabilities,Y_test):
    #Function returns Precision and recall by calculating for each threshold TP(true positives), FP(false positives), 
    #TN(true negatives) and FN (false negatives)
    
    
    global thresholds
    thresholds = np.linspace(0,1,20)
    precision_vector = np.zeros(len(thresholds))
    recall_vector = np.zeros(len(thresholds))
    False_Pos_rate_vector = np.zeros(len(thresholds))
    
    Y_test = np.array(Y_test) #converting to array
    
    for ind_threshold,value_threshold in enumerate(thresholds):
        if value_threshold < 1.0:
            Y_pred = np.where(probabilities>=value_threshold,1,0)
        else:
            Y_pred = np.where(probabilities>value_threshold,1,0)
        TP = 0
        FP = 0
        TN = 0
        FN = 0
        for i in range(len(Y_pred)):
            if Y_pred[i] == 1 and Y_test[i] == 1:
                TP += 1
            elif Y_pred[i] == 1 and Y_test[i] == 0:
                FP += 1
            elif Y_pred[i] == 0 and Y_test[i] == 1:
                FN += 1
            elif Y_pred[i] == 0 and Y_test[i] == 0:
                TN += 1
            
            
        False_pos_rate = FP/(TN + FP)
            
        try:
            recall = TP/(TP + FN)
        except ZeroDivisionError:
            recall = 0
            
        try:
            precision = TP/(TP + FP)
        except ZeroDivisionError:
            precision = 1.0
                
            
            
        precision_vector[ind_threshold] += precision
            
        recall_vector[ind_threshold] += recall
            
        False_Pos_rate_vector[ind_threshold] += False_pos_rate
    
    return precision_vector,recall_vector,False_Pos_rate_vector

# Function for getting scores given precision, recall and FPR
def get_performance(clf,metric_matrix_model, X_test, Y_test, metric):
    
    score = None
    
    metric_vectors = metric_matrix_model[:,0:3].astype(float)
    if metric == 'accuracy':
        score = round(clf.score(X_test, Y_test) * 100, 2)
    elif metric == 'prec':
        score = round(np.mean(metric_vectors[:,0])*100,2)
    elif metric == 'rec':
        score = round(np.mean(metric_vectors[:,1])*100,2)
    elif metric == 'auc':
        score = round(-1*np.trapz(metric_vectors[:,1],metric_vectors[:,2])*100,2)
        
    return score

def Evaluating_performance(clf_fitted,X_test,Y_test):
    # Evaluating performance given X and y
    
    scores = np.zeros((len(models),len(metrics)))
    metric_matrix = np.zeros((1,4)) #initializing the matrix for stacking performances arrays (precision,recall,FPR,model name)
    for clf_index in range(len(clf_fitted)):
        
        probabilities = clf_fitted[clf_index].predict_proba(X_test)[:,1]
        precision_vector,recall_vector,False_Pos_rate_vector = Confusion_matrix_precision_and_recall(probabilities,Y_test)
        
        model = np.repeat(names[clf_index],len(thresholds))
        metric_matrix_model = np.vstack([precision_vector,recall_vector,False_Pos_rate_vector,model]).T
        metric_matrix = np.vstack([metric_matrix,metric_matrix_model])
        
        for metric in range(len(metrics)):
             score = get_performance(clf_fitted[clf_index],metric_matrix_model,X_test, Y_test, metrics[metric])
             scores[clf_index][metric] = score
    return scores,metric_matrix[1::,:]


def Construct_performance_matrix(names,scores):
    # It builds dataframe of performances
    Compacted_perf_values = np.hsplit(scores,len(metrics)) 
    Performance_matrix = np.append(names,Compacted_perf_values).reshape(len(metrics)+1,len(models)).T     
    columns_vector = np.insert(metrics,0,"model") 
    Performance_matrix_df = pd.DataFrame(data=Performance_matrix,columns=columns_vector)
    return Performance_matrix_df

def Plotting_Roc_and_PR(metric_matrix):
    # Function plots ROC and PR curves 

    fig,(ax1,ax2) = plt.subplots(2,1,figsize=(12,9))

    for model in range(len(names)):
    
        matrix_list_string = metric_matrix[metric_matrix[:,3]==names[model],:][:,0:3] 
        matrix_list_num = matrix_list_string.astype(float)
        ax1.plot(matrix_list_num[:,2],matrix_list_num[:,1],linestyle='--', marker='o',lw = 2, label="AUC "+ names[model])
        ax1.plot([0,1],[0,1],linestyle='--',color="navy")
    
        ax2.plot(matrix_list_num[:,1],matrix_list_num[:,0],linestyle="--",marker='o',lw=2,label="PR "+ names[model])
    
        ax1.set_xlim([0.0, 1.0])
        ax1.set_ylim([0.0, 1.0])
    
        ax2.set_xlim([0.0, 1.0])
        ax2.set_ylim([0.0, 1.0])
    
        ax1.set_xlabel('False Positive Rate')
        ax1.set_ylabel('True Positive Rate')
    
        ax2.set_xlabel('Recall')
        ax2.set_ylabel('Precision')
    
        ax1.set_title('ROC curve')
        ax2.set_title('PR curve')
    
        ax1.legend(loc="lower right")
        ax2.legend(loc="lower right")
    
    plt.tight_layout()
    return fig
    
def Metrics_matrix_and_ROC_PR_plot_function(X_train,Y_train,X_test,Y_test,models,plot=False):
    #Main function which calls all other functions. It fits models, evaluates performances (precision, recall, roc_auc and
    # accuracy) and returns dataframe displaying performances and ROC, PR plots
    
    clf_fitted = []

    for model in range(len(models)):
        clf = models[model].fit(X_train,Y_train)
        clf_fitted.append(clf)
    
    scores,metric_matrix = Evaluating_performance(clf_fitted,X_test,Y_test)
    
    if plot == True:
        Performance_matrix_df = Construct_performance_matrix(names,scores)
        fig = Plotting_Roc_and_PR(metric_matrix)
        return Performance_matrix_df
    elif plot==False:
        Performance_matrix_df = Construct_performance_matrix(names,scores)
        return Performance_matrix_df
    
```

```python
df = pd.read_csv("titanic.csv") #Specify the path 
```

```python
#Preprocessing 

df['Sex'].replace(["female","male"],[0,1],inplace=True)
df['Embarked'].replace(["C","Q","S"],[0,1,2],inplace=True)
#Eliminating NAN values for our purpose
df_reduced = df.dropna(subset=["Age"],inplace=False)
df_reduced = df_reduced.dropna(subset=["Embarked"],inplace=False)
```

```python
#Preprocessing_2: Eliminating non Numerical Features
df_reduced.drop(labels=["Ticket","Cabin","Name"],axis=1,inplace=True)
```

```python
X =df_reduced.drop(labels=["Survived"],axis=1)
Y =df_reduced["Survived"]
#Reset indeces
X = X.reset_index(drop=True)
y = Y.reset_index(drop=True)
```

```python
# train test split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25)
print(X_train.shape)
print(X_test.shape)
```

```python
Metrics_matrix_and_ROC_PR_plot_function(X_train,y_train,X_test,y_test,models,plot=False)
```

```python
Metrics_matrix_and_ROC_PR_plot_function(X_train,y_train,X_test,y_test,models,plot=True)
```
